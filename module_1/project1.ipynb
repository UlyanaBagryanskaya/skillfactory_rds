```python
import pandas as pd
data = pd.read_csv('C:/Users/Xiaomi/Downloads/data.csv')
answer_ls = []

# первый вопрос - У какого фильма из списка самый большой бюджет?
data[(data.budget) == data.budget.max()].original_title 
answer_ls.append(4)

# второй вопрос - Какой из фильмов самый длительный (в минутах)?
data[(data.runtime) == data.runtime.max()].original_title 
answer_ls.append(2)

# третий вопрос - Какой из фильмов самый короткий (в минутах)?
data[(data.runtime) == data.runtime.min()].original_title 
answer_ls.append(3)

# четвертый вопрос - Какое число ближе к средней длительности фильма в датасете?
data.runtime.mean()
answer_ls.append(2)

# пятый вопрос - Какое число ближе к медианной длительности фильма в датасете?
data.runtime.median()
answer_ls.append(1)

# шестой вопрос - Какой самый прибыльный фильм?
data_add = pd.DataFrame()
data_add['profit'] = data['revenue'] - data['budget']
data_all = pd.concat([data, data_add], axis = 1)
data_all[(data_all.profit) == data_all.profit.max()].original_title #выводим название 
answer_ls.append(5)

# седьмой вопрос - Какой фильм самый убыточный? - аналогично предыдущему
data_all[(data_all.profit) == data_all.profit.min()].original_title #выводим название 
answer_ls.append(2)

# восьмой вопрос - У скольких фильмов из датасета объем сборов оказался выше бюджета?
data_all[(data_all.profit) > 0].count()
answer_ls.append(1)

# девятый вопрос - Какой фильм оказался самым кассовым в 2008 году?
data_all[((data_all.release_year) == 2008)]['profit'].max()
data_all[(data_all.profit == 816921825)].original_title
answer_ls.append(4)

# десятый вопрос - Самый убыточный фильм за период с 2012 по 2014 гг. (включительно)?
data_0 = data_all.query('release_year in [2012,2013,2014]')
data_0[(data_0.profit) == data_0.profit.min()].original_title
answer_ls.append(5)

#одиннадцатый вопрос - Какого жанра фильмов больше всего?
pd.DataFrame(data.genres.str.split('|').tolist()).stack().value_counts()
answer_ls.append(3)

# двенадцатый вопрос - Какого жанра среди прибыльных фильмов больше всего?
# для начала сделаем табличку всех прибыльных фильмов - тех, где разница между сборами и бюджетом больше нуля
pro_films = data_all.query('profit > 0')
pd.DataFrame(pro_films.genres.str.split('|').tolist()).stack().value_counts()
answer_ls.append(3)

# тринадцатый вопрос - Кто из режиссеров снял больше всего фильмов?
data.director.value_counts()
answer_ls.append(3)

#четырнадцатый вопрос - Кто из режиссеров снял больше всего прибыльных фильмов?
pro_films = data_all.query('profit > 0')
pro_films
pro_films.director.value_counts()
answer_ls.append(2)

# пятнадцатый вопрос - кто из режиссеров принес больше всего прибыли
pro_films.groupby(['director'])['profit'].sum().sort_values(ascending=False)
dir = pd.DataFrame(data_all.director.str.split('|').tolist()).stack().reset_index() 
dir.columns =  ['row1', 'row2', 'director']
dir['margin_dir'] = dir.row1.apply(lambda x: data_all.iloc[x]['profit'])
dir.groupby(['director'])['margin_dir'].sum().sort_values(ascending=False)
answer_ls.append(5)

# шестнадцатый вопрос - какой актер принес больше всего прибыли
actors = pd.DataFrame(data_all.cast.str.split('|').tolist()).stack().reset_index()
actors.columns = ['row3', 'row4', 'cast']
actors.head()
actors['profit_act'] = actors.row3.apply(lambda x: data_all.iloc[x]['profit'])
actors.groupby(['cast'])['profit_act'].sum().sort_values(ascending = False)
answer_ls.append(1)

# семнадцатый вопрос -Какой актер принес меньше всего прибыли в 2012 году?
films_2012 = data_all.query('release_year == 2012')
act_min = pd.DataFrame(films_2012.cast.str.split('|').tolist()).stack().reset_index()
act_min.columns = ['row3', 'row4', 'cast']
act_min
act_min['profit_act'] = act_min.row3.apply(lambda x: films_2012.iloc[x]['profit'])
act_min.groupby(['cast'])['profit_act'].sum().sort_values()               
answer_ls.append(3)

# восемнадцатый вопрос - Какой актер снялся в большем количестве высокобюджетных фильмов?
high_bud = data[(data.budget > data.budget.mean())]
actors2 = pd.DataFrame(high_bud.cast.str.split('|').tolist()).stack().reset_index()
actors2.columns = ['row5', 'row6', 'name']
actors2['name'].value_counts()
answer_ls.append(3)

# девятнадцатый вопрос - В фильмах какого жанра больше всего снимался Nicolas Cage?
NCfilms = data[data.cast.str.contains("Nicolas Cage", na=False)]
pd.DataFrame(NCfilms.genres.str.split('|').tolist()).stack().value_counts()
answer_ls.append(2)

# двадцатый вопрос - Какая студия сняла больше всего фильмов?
pd.DataFrame(data.production_companies.str.split('|').tolist()).stack().value_counts()
answer_ls.append(1)

# Вопрос 21 - Какая студия сняла больше всего фильмов в 2015 году?
films2015 = data.query('release_year == 2015')
pd.DataFrame(films2015.production_companies.str.split('|').tolist()).stack().value_counts()
answer_ls.append(4)

# Вопрос 22 - Какая студия заработала больше всего денег в жанре комедий за все время?
ComFilms1 = data_all[data_all.genres.str.contains('Comedy', na = False)]
ComFilms = pd.DataFrame(ComFilms1.production_companies.str.split('|').tolist()).stack().reset_index()
ComFilms.columns = ['row7', 'row8', 'company_name']
ComFilms
ComFilms['profit_comp'] = ComFilms.row7.apply(lambda x: ComFilms1.iloc[x]['profit'])
ComFilms.groupby(['company_name'])['profit_comp'].sum().sort_values(ascending = False)
answer_ls.append(2)

# Вопрос 23 - Какая студия заработала больше всего денег в 2012 году?
films2012 = data_all.query('release_year == 2012')
ProdComp = pd.DataFrame(films2012.production_companies.str.split('|').tolist()).stack().reset_index()
ProdComp.columns = ['row9', 'row10', 'company_name']
ProdComp['profit_comp'] = ProdComp.row9.apply(lambda x: films2012.iloc[x]['profit'])
ProdComp.groupby(['company_name'])['profit_comp'].sum().sort_values(ascending = False)
answer_ls.append(3)

# Вопрос 24 - Самый убыточный фильм от Paramount Pictures?
parmfilms = data_all[data_all.production_companies.str.contains("Paramount Pictures", na=False)]
parmfilms[(parmfilms.profit) == parmfilms.profit.min()].original_title
answer_ls.append(1)

# Вопрос 25 - Какой самый прибыльный год (в какой год студии заработали больше всего)?
data_all.groupby(['release_year'])['profit'].sum().sort_values(ascending = False)
answer_ls.append(5)

# Вопрос 26 - Какой самый прибыльный год для студии Warner Bros?
WBfilms = data_all[data_all.production_companies.str.contains("Warner Bros", na=False)]
WBfilms.groupby(['release_year'])['profit'].sum().sort_values(ascending = False)
answer_ls.append(1)

# Вопрос 27 - В каком месяце за все годы суммарно вышло больше всего фильмов?
data['month'] = [release_date.split('/')[0] for release_date in data.release_date]
data.month.sort_values(ascending = False)
answer_ls.append(4)

# Вопрос 28 - Сколько суммарно вышло фильмов летом (за июнь, июль, август)?
summer_data = data[(data.month == '6') | (data.month == '7') | (data.month == '8')]
len(summer_data)
answer_ls.append(2)

# Вопрос 29 - Какой режиссер выпускает (суммарно по годам) больше всего фильмов зимой?
winter_data = data[(data.month == '12') | (data.month == '1') | (data.month == '2')]
winter_data.groupby(['director'])['month'].count().sort_values(ascending = False)
answer_ls.append(5)

# Вопрос 30 - Какой месяц чаще всего самый прибыльный в году?
#data_all['month'] = pd.to_datetime(data_all.release_date).apply(lambda x: x.month)
#pivot = data_all.pivot_table(columns = 'release_year', index = 'month', values = 'profit', aggfunc = 'sum')
#pivot.idxmax().value_counts()
answer_ls.append(2)

# Вопрос 31 - Названия фильмов какой студии в среднем самые длинные по количеству символов?
data_all['lenght'] = data.original_title.apply(lambda x:len(x))
ProdComp = pd.DataFrame(data_all.production_companies.str.split('|').tolist()).stack().reset_index()
ProdComp.columns = ['row11', 'row12', 'company_name']
ProdComp['comp_len'] = ProdComp.row11.apply(lambda x: data_all.iloc[x]['lenght'])
ProdComp.groupby(['company_name'])['comp_len'].mean().sort_values(ascending = False)
answer_ls.append(5)

# Вопрос 32 - Названия фильмов какой студии в среднем самые длинные по количеству слов?
data_all['word_len'] = data.original_title.apply(lambda x: len(x.split()))
ProdComp['comp_wl'] = ProdComp.row11.apply(lambda x: data_all.iloc[x]['word_len'])
ProdComp.groupby(['company_name'])['comp_wl'].mean().sort_values(ascending = False)
answer_ls.append(5)

#Вопрос 33 - Сколько разных слов используется в названиях фильмов (без учета регистра)?
low_titles = data.original_title.apply(lambda x: [i.lower() for i in x.split(' ')]).tolist()
pd.DataFrame(low_titles).stack().nunique()
answer_ls.append(3)

#Вопрос 34 - Какие фильмы входят в 1 % лучших по рейтингу?
data.sort_values(by='vote_average', ascending=False).head(int(len(data)*0.01))['original_title']
answer_ls.append(1)

#Вопрос 35 - Какие актеры чаще всего снимаются в одном фильме вместе?
import itertools
import collections
act2 = collections.Counter()
for i, k in data.iterrows():
    actor = k.cast.split('|')
    act2.update([ x for x in itertools.combinations(actor, 2)])
act2.most_common()
answer_ls.append(5)

# Вопрос 36 - У какого из режиссеров самый высокий процент фильмов со сборами выше бюджета?
#сколько всего у одного реж фильмов, потом сколько из них выше бюджета, второе делить на первое, искать максимум

#directors = {}
#for i, k in data.iterrows():
#    director = k.director.split('|')
#    if k.revenue > k.budget:
#        for direct in director:
#            if direct not in directors.keys():
#                directors[direct] = 1
#            else:
#                directors[direct] += 1

#max_val = max(directors.values())
#final_dict = {k:v for k, v in directors.items() if v == max_val}
#final_dict - получился {'Ridley Scott': 12}
#QTfilms = data_all[data_all.director.str.contains("Quentin Tarantino", na=False)]
#QTfilms # 6 из 7 фильмов прибыльные
#SSfilms = data_all[data_all.director.str.contains("Steven Soderbergh", na=False)]
#SSfilms # 10 из 13
#RRfilms = data_all[data_all.director.str.contains("Robert Rodriguez", na=False)]
#RRfilms # 9 из 11
CNfilms = data_all[data_all.director.str.contains("Christopher Nolan", na=False)]
CNfilms # 100% 
#CEfilms = data_all[data_all.director.str.contains("Clint Eastwood", na=False)]
#CEfilms # 10 из 12

answer_ls.append(4)

pd.DataFrame({'Id':range(1,len(answer_ls)+1), 'Answer':answer_ls}, columns=['Id', 'Answer'])
```
